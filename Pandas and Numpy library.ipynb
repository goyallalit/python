# cross tab
# difference b/w cross tab and pivot 
# The function does not require a dataframe as an input. 
# It can also accept array-like objects for its rows and columns.
# The function can normalize the resulting dataframe, meaning that the values displayed can be displayed as percentage 
# of row or column totals.
# The default function is len (count), whereas the pivot table one is numpy’s mean function.

x=pd.crosstab(index= df.MaritalStatus  ,
              columns= df.Gender  ,
               values= df.Income  ,
               aggfunc   ='sum')
print(x)
#A key benefit of the crosstab function over the Pandas Pivot Table function is that it allows you to normalize the resulting 
#dataframe, returning values displayed as percentages.
#This goes one step further – the normalize argument accepts a number of different options:

#‘all’ or True – normalizes the values across the entire dataframe (as a percentage of the total across rows and columns)
#‘index’ – normalizes across rows
#‘columns’ – normalizes down columns
x1=pd.crosstab(index= df.MaritalStatus  ,
              columns= df.Gender  
              )
#Pandas will generate a crosstab which counts the number of times each item appears (the length of that series)
print(x1)
x2=pd.crosstab(index= df.MaritalStatus  ,
              columns= df.Gender  ,normalize= 'index'
              )
print(x2)
